## 1 - Importações:

```python
import numpy as np
import matplotlib.pyplot as plt
from MRPy import MRPy
```

## 2 - Sistema em Ressonância:

```python
#Definindo as propriedades do sistema:

m = 1.0          # massa do sistema (kg)
fn = 10        # frequência natural de vibração (Hz)
wn = fn*np.pi*2  # frequência natural de vibração (rad/s)
zt = 0.02        # Razão de amortecimento do sistema


Td = 5         # tempo total de integração
fs = 2000        # frequência de amostragem

# vetor com os tempos:
t = np.linspace(0,Td,Td*fs)
F = MRPy.zeros(1,Td*fs,fs=fs)

F[0,:] = 1*np.sin(wn*t)
u = F.sdof_Duhamel(fn,zt)/m
u.plot_time()
F.plot_time()
```
