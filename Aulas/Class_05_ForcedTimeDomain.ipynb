## 2 - Aproximação da Resposta Impulsiva:

```python
F0 = 0.40              # Força Constante (N)
td = 0.25              # Duração da Carga (s)
Td = 20                # Tempo total da integração
fs = 2096              # Número de pontos por segundo

F = MRPy.zeros(1, Td*fs, fs=fs)

i0 = int(td*F.fs)      # Número de pontos até chegar em td
F[0, 0:i0] = F0

F.plot_time(figsize=(9,3))
```

```python
m = 1.0             # massa do sistema em kg
fn = 2              # Frequência natural do sistema
zt = 0.03           # Razão de amortecimento em relação ao crítico

u = F.sdof_Duhamel(fn,zt)/m
u.plot_time()
```
