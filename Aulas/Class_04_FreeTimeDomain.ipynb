## 1- Importing Libraries:

## 2 - Simple System Vibrating:

```python
fn = 0.5          # System Natural Frequency (Hz)
wn = 2*np.pi*fn   # System Natural Frequency (rads/s)
u0 = 1            # Initial Displacement (m)
v0 = 30           # Initial Speed (m/s)

Td = 20           # Time Series Duration (s)
fs = fn*100         # Sampling Rate (Hz)

t = np.linspace(0,Td, int(fs*Td)) #time vector
u = u0*np.cos(wn*t) + (v0/wn)*np.sin(wn*t)

plt.figure(1, figsize=(10,5), clear=True)
plt.plot(t,u)

plt.xlim(0,Td)
plt.xlabel('Time (s)')

plt.ylim(-10,10)
plt.ylabel('u(t)')
```

### 3 - Damped System:

```python
fn = 0.5          # System Natural Frequency (Hz)
wn = 2*np.pi*fn   # System Natural Frequency (rads/s)
u0 = 10            # Initial Displacement (m)
v0 = 0           # Inicial Speed (m/s)
zt = 0.03         # Damping ratio to critical

Td = 20           # Time Series Duration (s)
fs = fn*100       # Sampling Rate (Hz)

t = np.linspace(0, Td, int(fs*Td))

wD = wn*np.sqrt(1 - zt**2)
e = np.exp(-zt*wn*t)
uD = e*(u0*np.cos(wD*t) + ((v0 + zt*wn*u0)/wD)*np.sin(wD*t)) 

plt.figure(2, figsize=(8, 4), clear=True)
plt.plot(t, uD)
plt.plot(t, u0*e, 'r:', t, -u0*e, 'r:')

plt.xlim(  0, Td);  plt.xlabel('time (s)') 
plt.ylim(-10, 10);  plt.ylabel('u(t)') 

plt.grid(True) 
```
